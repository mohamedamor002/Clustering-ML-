```python
import pandas as pd 
df = pd.read_csv(r"/content/CC GENERAL.csv")
```

```python
df=df.drop("CUST_ID", axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df["CREDIT_LIMIT"].median(),inplace = True)  
df.fillna(df["MINIMUM_PAYMENTS"].median(),inplace = True) 
df
```

```python
from sklearn.preprocessing import StandardScaler
def scaling(data):

  #Initalise standard scaler
  scaler = StandardScaler()
  #Fit the scaler using X_train data
  scaler.fit(data)
  #Transform X_train and X_test using the scaler and convert back to DataFrame
  data = pd.DataFrame(scaler.transform(data), columns = data.columns)
  return data
```

```python
df = scaling(df)
df
```

```python
from sklearn.cluster import KMeans 
kmeans=KMeans(n_clusters=5, random_state=0)
kmeans.fit(df)
identified_clusters = kmeans.predict(df) #THIS IS WHAT GIVE YOU LABELS

l = []
#l= kmeans.labels() 
labels = pd.DataFrame(identified_clusters)
labels
df['labels_'] = labels
```

```python
labels.value_counts()
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(df["PAYMENTS"][df.labels_ == 0],          
            df["CREDIT_LIMIT"][df.labels_ == 0],s=80,c='magenta',label='Careful')
plt.scatter(df["PAYMENTS"][df.labels_ == 1],
           df["CREDIT_LIMIT"][df.labels_ == 1],s=80,c='yellow',label='Standard')
plt.scatter(df["PAYMENTS"][df.labels_ == 2],
           df["CREDIT_LIMIT"][df.labels_ == 2],s=80,c='green',label='Target')
plt.scatter(df["PAYMENTS"][df.labels_ == 3], 
           df["CREDIT_LIMIT"][df.labels_ == 3],s=80,c='cyan',label='Careless')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Payments')
plt.ylabel('Credit Limit')
plt.legend()
plt.show()
```

```python

Sum_of_squared_distances = []

K = range(1,30)
for k in K :
    km = KMeans(n_clusters = k)
    km = km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
Sum_of_squared_distances
plt.plot(K, Sum_of_squared_distances, "bx-")
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title("The Elbow Method for Optimal K")
plt.show()
```
